Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и

````python
import numpy as np
A = [[(j + i + 1) % 2 for j in range(8)] for i in range(8)]
for i in A:
    print(i)
````

# 3.3.2
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.

````python
import pandas as pd
from numpy import min, max, mean, std

# Загрузка данных
url = "https://raw.githubusercontent.com/akmand/datasets/master/iris.csv"
df = pd.read_csv(url)

# Извлекаем нужные признаки как массивы NumPy
sepal_length = df['sepal_length_cm'].values
sepal_width = df['sepal_width_cm'].values

# Минимаксная нормализация для sepal_length: (x - min) / (max - min)
sepal_length_norm = (sepal_length - min(sepal_length)) / (max(sepal_length) - min(sepal_length))

# Z-масштабирование для sepal_width: (x - mean) / std
sepal_width_scaled = (sepal_width - mean(sepal_width)) / std(sepal_width)

# Выводим первые 5 значений для проверки
print("Нормализованный sepal_length (минимакс):")
print(sepal_length_norm )

print("\nМасштабированный sepal_width (z-score):")
print(sepal_width_scaled)
````
