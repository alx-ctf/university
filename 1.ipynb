````python
#4.3.1
import numpy as np
import matplotlib.pyplot as plt
from random import random

array = np.random.rand(10) # массив 10 чисел от 0 до 1
mean = np.mean(array) # средняя
median = np.median(array) # медиана

x = np.arange(10)

plt.ylim(0, 1) # ограничение по y
plt.xlim(0, 10) # ограничение по x
plt.scatter(x, array, color='#FF0000', s=10, label='Случайные числа')
plt.axhline(y=mean, color='#0000FF', linestyle='-', label=f'Среднее: {mean:.3f}')
plt.axhline(y=median, color='#00FF00', linestyle='-.', label=f'Медиана: {median:.3f}')

plt.xlabel('Индекс')
plt.ylabel('Значение')
plt.title('Точечная диаграмма случайных чисел') # заголовок
plt.legend()
plt.show()
````

````python
#4.3.2
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (np.sqrt(1 + np.exp(x) + np.cos(x)**2) / np.abs(1 - np.sin(x)) + np.log(np.abs(2 * x)))


x = np.linspace(1, 10, 100) # 100 точек от 1 до 10
y = f(x)

#Выделение среза первой половины массива
x_1half = x[:len(x)//2: ]
y_1half = y[:len(x)//2: ]
x_2half = x[ len(x)//2: ]
y_2half = y[ len(x)//2::]

plt.plot(x_1half, y_1half, '.', label='Первая половина')
plt.plot(x_2half, y_2half, '-', label='Вторая половина')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Функция и срез первой половины')
plt.legend()
plt.grid(True) #сетка

plt.show()
````

````python
#4.3.3
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simpson
from numpy import trapezoid

def f(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 10.1, 1)  # Включаем 0 и 10 с шагом 1
y = f(x)

# Вычисление площади под графиком
area_trapz = trapezoid(y, x) # компилятор ругается на trapz
area_simps = simpson(y, x) # в scipy нет simps но есть simpson


plt.plot(x, y, 'r-', label='f(x)')
plt.fill_between(x, y, alpha=0.5, label='Площадь под кривой')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции и площадь под ним')
plt.legend()

plt.text(2, 0.8, f'Площадь (трапеции): {area_trapz:.4f}', bbox=dict(facecolor='white', alpha=0.8))
plt.text(2, 0.7, f'Площадь (Simpson): {area_simps:.4f}', bbox=dict(facecolor='white', alpha=0.8))

plt.grid(True)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes


dates = ['01.01.2021', '01.02.2021', '01.03.2021', '01.04.2021', '01.05.2021', '01.06.2021', '01.07.2021', '01.08.2021', '01.09.2021', '01.10.2021', '01.11.2021', '01.12.2021']

apple_inc    = [131.96, 121.26, 122.15, 131.46, 124.61, 136.96, 145.86, 151.83, 141.50, 149.80, 165.30, 177.57]
alphabet_inc = [91.79,  101.84, 103.43, 120.51, 120.58, 125.32, 135.22, 145.46, 133.27, 148.27, 142.45, 144.68]
microsoft    = [231.96, 232.38, 235.77, 252.18, 249.68, 270.90, 284.91, 301.88, 281.92, 331.62, 330.59, 336.32]

plt.figure(figsize=(12, 6)) # 12 на 6
plt.plot(dates, apple_inc,    label='Apple', marker='o', color='blue')
plt.plot(dates, alphabet_inc, label='google', marker='o', color='red')
plt.plot(dates, microsoft,    label='microsoft', marker='o', color='green')

plt.title('Процентный рост и падение акций компаний')
plt.xlabel('Дата')
plt.ylabel('Процент изменения (%)')
plt.legend() # квадратик справа сверху
plt.grid() # сетка
plt.axhline(apple_inc[0], color='blue', linewidth=0.7, linestyle='--')
plt.axhline(alphabet_inc[0], color='red', linewidth=0.7, linestyle='--')
plt.axhline(microsoft[0], color='green', linewidth=0.7, linestyle='--')

# Установка шага на оси Y в 100
plt.yticks(np.arange(min(min(apple_inc), min(alphabet_inc), min(microsoft)),
                     max(max(apple_inc), max(alphabet_inc), max(microsoft)) + 100, 100))

# квадратик слева сверху
apple_change = ((apple_inc[-1] - apple_inc[0]) / apple_inc[0]) * 100
alphabet_change = ((alphabet_inc[-1] - alphabet_inc[0]) / alphabet_inc[0]) * 100
microsoft_change = ((microsoft[-1] - microsoft[0]) / microsoft[0]) * 100

inset_ax = inset_axes(plt.gca(), width="17%", height="12%", loc='upper left')
inset_ax.text(0.1, 0.7, f'Apple: +{apple_change:.1f}%', fontsize=10, color='blue')
inset_ax.text(0.1, 0.4, f'Google: +{alphabet_change:.1f}%', fontsize=10, color='red')
inset_ax.text(0.1, 0.1, f'Microsoft: +{microsoft_change:.1f}%', fontsize=10, color='green')
inset_ax.set_xticks([])
inset_ax.set_yticks([])

plt.show()
````
