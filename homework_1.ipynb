````python
start, end, step = 1, 10, 3 # –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏ –≤—ã–Ω–µ—Å –Ω–∞—á–∞–ª–æ –∫–æ–Ω–µ—Ü –∏ —à–∞–≥

x = start
while x <= end: # —Å –Ω–∞—á–∞–ª–∞ –¥–æ –∫–æ–Ω—Ü–∞
    print(x, end=" ")
    x+=step # x = x + step   |+|

print()
x = end
while x>=start: # —Å –∫–æ–Ω—Ü–∞ –¥–æ –Ω–∞—á–∞–ª–∞
    print(x, end=" ")
    x-=step # x = x - step   |-|
````

# 4.3.2
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è F –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π

````python
from numpy import arange, sqrt, exp, cos, log, abs, sin
from matplotlib.pyplot import legend, figure, plot, scatter, title, xlabel, ylabel, grid, show
x = arange(1, 11)
n = 5
numerator = sqrt(1 + exp(sqrt(x)) + cos(x**2)) + log(abs(2 * x))
denominator = abs(1 - sin(x)**3)    
f = numerator / denominator
x_slice = x[:n]
f_slice = f[:n]
figure(figsize=(10, 6))
plot(x, f, marker='o', linestyle='-', color='gray', linewidth=1, markersize=5, label='–í–µ—Å—å –º–∞—Å—Å–∏–≤')
scatter(x_slice, f_slice, color='red', s=100, zorder=5, label='–°—Ä–µ–∑ (1-—è –ø–æ–ª–æ–≤–∏–Ω–∞)', edgecolor='black')
title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Å –≤—ã–¥–µ–ª–µ–Ω–Ω—ã–º —Å—Ä–µ–∑–æ–º')
xlabel('x')
ylabel('f(x)')
grid(True, alpha=0.4)
legend()
show()
````

# 4.3.3
–î–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏—è F2 –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞
–∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å
–ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(),
–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.
–î–ª—è –µ–µ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–π —Ä–∞–±–æ—Ç—ã –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø–æ–¥–∫–ª—é—á–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ
–±–∏–±–ª–∏–æ—Ç–µ–∫–∏:
```py
from scipy.integrate import simps
from numpy import trapz
```

````python
#4.3.3
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simpson
from numpy import trapezoid

def f(x):
    return np.abs(np.cos(x * np.exp(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 10.1, 1)  # –í–∫–ª—é—á–∞–µ–º 0 –∏ 10 —Å —à–∞–≥–æ–º 1
y = f(x)

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –ø–ª–æ—â–∞–¥–∏ –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º
area_trapz = trapezoid(y, x) # –∫–æ–º–ø–∏–ª—è—Ç–æ—Ä —Ä—É–≥–∞–µ—Ç—Å—è –Ω–∞ trapz
area_simps = simpson(y, x) # –≤ scipy –Ω–µ—Ç simps –Ω–æ –µ—Å—Ç—å simpson


plt.plot(x, y, 'r-', label='f(x)')
plt.fill_between(x, y, alpha=0.5, label='–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –∫—Ä–∏–≤–æ–π')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –∏ –ø–ª–æ—â–∞–¥—å –ø–æ–¥ –Ω–∏–º')
plt.legend()

plt.text(2, 0.8, f'–ü–ª–æ—â–∞–¥—å (—Ç—Ä–∞–ø–µ—Ü–∏–∏): {area_trapz:.4f}', bbox=dict(facecolor='white', alpha=0.8))
plt.text(2, 0.7, f'–ü–ª–æ—â–∞–¥—å (Simpson): {area_simps:.4f}', bbox=dict(facecolor='white', alpha=0.8))

plt.grid(True)
plt.show()
````

# 4.3.5
–°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è
–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤ + ‚Ä≤, ‚Ä≤ ‚àí ‚Ä≤, ‚Ä≤ ‚àô ‚Ä≤, ‚Ä≤ √∑ ‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ
—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: F3, F4, F5, F6

````python
import math

def right_input(str_ : str) -> int:
    while True:
        number = input(f"{str_} >> ")
        if not number.isdecimal():
            print("not decimal")
            continue
        else:
            return int(number)

def option_match(x : int, y: int):
  print_text()
  while True:
    option = right_input("enter option")
    match option:
      case 1:
        print(x+y)
        break
      case 2:
        print(x-y)
        break
      case 3:
        print(x*y)
        break
      case 4:
        print(x/y)
        break
      case 5:
        print(f"{math.exp(x+y):.3f}")
        break
      case 6:
        print(f"{math.sin(x+y):.3f}")
        break
      case 7:
        print(f"{math.cos(x+y):.3f}")
        break
      case 8:
        print(f"{math.pow(x, y):.3f}")
        break
      case _:
        print("invalid option")

def print_text():
  print("enter option:\n1. +\n2. -\n3. *\n4. /\n5. exp\n6. sin\n7. cos\n8. ^")


if __name__ == "__main__":
  x = right_input("enter first number")
  y = right_input("enter second number")
  option_match(x, y)
````
